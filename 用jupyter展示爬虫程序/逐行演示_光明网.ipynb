#### 根据日期，生成对应日期的新闻界面

```python
date = 20210202
#https://epaper.gmw.cn/gmrb/html/2021-03/10/nw.D110000gmrb_20210310_1-01.htm
#https://epaper.gmw.cn/gmrb/html/2021-03/09/nw.D110000gmrb_20210309_1-01.htm

# 网址中段需要根据输入的年月日进行转换    
# 将输入的数字转化为字符串，以供组成网址
year = str(date)[:4]    
# 小于10的月份格式为0+月份，大于等于10的月份不变。如5月为05,11月为11
month = str(date)[4:6]  
# 小于10的日期格式为0+日期，大于等于10的日期不变。如3号为03,15号为15
day = str(date)[6:]  

# 根据年月日生成对应的网址
url = f'https://epaper.gmw.cn/gmrb/html/{year}-{month}/{day}/nw.D110000gmrb_{year}{month}{day}_1-01.htm'
print(url)
```

#### 分析该网页

```python
# 伪装成chrome浏览器进行网页登录
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36'}
# 文件保存地址
dirs_save = './爬取的文件/'

# 从指定网页url的内容读入到webdata中，以供分析使用
webdata = requests.get(url, headers=headers)
print(webdata.text)
# 通过xpath模块来分析网页内容
selector = etree.HTML(webdata.text)
```

#### 获得新闻标题

```python
titles = selector.xpath('//*[@class="text_c"]/h1/text()')
#titles = selector.xpath('//*[@class="list_t"]/div/h1/text()')
title = ''
# 标题可能为两行
for line in titles:
    title += line
print(title)
```

#### 获得新闻内容

```python
text_list = selector.xpath('//*[@id="articleContent"]/p/text()')
print(text_list)
```

#### 获得图片（的网址）

```python
img_url_part_list = selector.xpath('//*[@class="imgBox"]/img/@src')
#img_url_part_list = selector.xpath('//*[@class="c_c"]/div/img/@src')
print(img_url_part_list)
img_url_list = []
for url_part in img_url_part_list:
    img_url = url + '/../' + url_part
    img_url_list.append(img_url)
    print(img_url)
```

#### 对爬取的数据进行保存

##### --将标题设置为可保存的文件名

```python
"""处理文件名非法字符的方法"""
title = title.replace(' ','-')
pattern = re.compile(r"[\/\\\:\*\?\"\<\>\|]")  # '/ \ : * ? " < > |'
new_title = re.sub(pattern, "_", title)  # 替换为下划线
print(new_title)
# 文件名为保存路径+标题名称
txt_name = dirs_save + f'{new_title}.txt'
print(txt_name)
```

##### --保存新闻内容

```python
f = open(txt_name, 'w', encoding = 'utf-8')
for text in text_list:
    f.write(text)    
print(f'已保存  {txt_name}') 
f.close()
```

##### --保存图片

```python
img_num = 1
for img_url in img_url_list:
    
    # 根据图片地址，获取图片数据
    imgdata = requests.get(img_url, headers=headers).content
    
    
    # 为多张图片分别命名
    save_name = dirs_save + f'{new_title}_{img_num}.jpg'
    img_num +=1
    # 写入图片
    f = open(save_name,'wb')
    f.write(imgdata)
    print(f'已保存  {save_name}')
    f.close()
```
